#Questão 2 - até 2 pontos
Usando apenas álgebra, encontre os pontos críticos das funções a seguir determinando a natureza destes pontos críticos, caso eles existam:

a) $f(x)=x^4-4x^3+4x^2$

b)$f(x)=x^4(x-3)^2$</br>


**Rubrica de Avaliação**: até dois pontos pela identicação correta da natureza dos pontos críticos, caso eles existam, ou não. Caso um ponto crítico seja identificado de forma errada ele anulará o ponto correpondente ao subitem (a ou b) da questão.



### A)

**Função:**
$$ f(x) = x^4 - 4x^3 + 4x^2 $$

**Passo 1: Derivada da função**
$$ f'(x) = 4x^3 - 12x^2 + 8x $$

**Passo 2: Encontrar os pontos críticos**
$$ 4x(x^2 - 3x + 2) = 0 $$
$$ x(x-1)(x-2) = 0 $$
Os pontos críticos são \( x = 0 \), \( x = 1 \) e \( x = 2 \).

**Passo 3: Determinar a natureza dos pontos críticos usando o teste da segunda derivada:**
$$ f''(x) = 12x^2 - 24x + 8 $$
$$ f''(0) = 8 > 0 \quad \text{(mínimo local)} $$
$$ f''(1) = -4 < 0 \quad \text{(máximo local)} $$
$$ f''(2) = 8 > 0 \quad \text{(mínimo local)} $$

Assim, os pontos críticos \( x = 0 \) e \( x = 2 \) são mínimos locais, e \( x = 1 \) é um máximo local.

### B)

**Função:**
$$ f(x) = x^4(x-3)^2 $$

**Passo 1: Derivada da funçao**
$$ f'(x) = 4x^3(x-3)^2 + 2x^4(x-3) $$
$$ f'(x) = 2x^3(x-3)(2x(x-3) + x) $$
$$ f'(x) = 2x^3(x-3)(3x - 3) $$
$$ f'(x) = 6x^3(x-3)(x - 1) $$

**Passo 2: Encontrar os pontos críticos**
$$ 6x^3(x-3)(x - 1) = 0 $$
Os pontos críticos sao \( x = 0 \), \( x = 3 \) e \( x = 1 \).

**Passo 3: Determinar a natureza dos pontos críticos usando o teste da segunda derivada:**
$$ f''(x) = 18x^2(x-3)(x-1) + 6x^3(2x - 4) $$
$$ f''(x) = 6x^2(x-3)(5x - 3) $$
$$ f''(0) = 0 \quad \text{(inconclusivo)} $$
$$ f''(3) = 0 \quad \text{(inconclusivo)} $$

Vamos utilizar a terceira derivada para mais informaçao:
$$ f'''(x) = 6x(x-3)(10x-3) + 12x^2(5x-3) $$
$$ f'''(0) \neq 0 $$
$$ f'''(3) \neq 0 $$

Estes resultados sugerem que $$ x = 0 $$ $$ x = 3 $$ são pontos de inflexão.

Concluímos que:
- Para a função A, os pontos  (x = 0) e (x = 2) são mínimos locais, e (x = 1) é um máximo local.
- Para a função B, (x = 0) e (x = 3) são pontos de inflexão.



# Questão 3 - Vale até 4 Pontos

Usando apenas ferramentas computacionais e gráficos, faça a análise dos pontos críticos das seguintes funções:

a)$f(x)=e^{2x}−e^x$

b)$f(x)=e^x+x$

**Atenção**: O tempo é importante, então escolha, entre usar ferramentas computacionais ou gráficas. Independente da sua escolha você deverá escrever a resposta em uma célula de textos.


```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função e suas derivadas
def f(x):
    return np.exp(2 * x) - np.exp(x)

def f_prime(x):
    return 2 * np.exp(2 * x) - np.exp(x)

def f_double_prime(x):
    return 4 * np.exp(2 * x) - np.exp(x)

# Intervalo para o gráfico
x = np.linspace(-2, 1, 400)

# Calculando os valores da função e das derivadas
y = f(x)
y_prime = f_prime(x)
y_double_prime = f_double_prime(x)

# Encontrando o ponto crítico
critical_point = -np.log(2)

# Plotando
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = $e^{2x} - e^x$')
plt.plot(x, y_prime, label="f'(x)", linestyle='--')
plt.scatter([critical_point], [f(critical_point)], color='red', zorder=5)  # Ponto crítico
plt.title('Análise de Pontos Críticos da Função $e^{2x} - e^x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo a função e suas derivadas
def f(x):
    return np.exp(x) + x

def f_prime(x):
    return np.exp(x) + 1

# Intervalo para o gráfico
x = np.linspace(-2, 2, 400)

# Calculando os valores da função e das derivadas
y = f(x)
y_prime = f_prime(x)

# Plotando
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='f(x) = $e^x + x$')
plt.plot(x, y_prime, label="f'(x)", linestyle='--')
plt.title('Análise de Pontos Críticos da Função $e^x + x$')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.grid(True)
plt.show()
```

Para a função $f(x) = e^{2x} - e^x$, o ponto crítico encontrado é $x = -\ln(2)$. A análise da segunda derivada mostra que este é um ponto de máximo local, pois $f''(x) < 0$ neste ponto.

#### Função b) $f(x) = e^x + x$

A função $f(x) = e^x + x$ não possui pontos críticos, pois a derivada $f'(x) = e^x + 1$ nunca é zero. A função é estritamente crescente em todo o domínio real.


# Questão 4 - Alternativa - Vale até 2 pontos

Em uma fábrica, latas de sopa cilíndricas devem ser fabricadas para conter um volume dado $V$. Não há desperdício de material na corte da parte vertical de cada lata ( o lado do cilíndro), mas cada tampa e fundo, que são círculos de raio $r$, são cortados de um quadrado que mede $2r$ unidades de lado. Assim, o material usado para fabricar cada lata de sopa tem uma área de $A = 2 \pi rh + 8r^2$ unidades quadradas. Encontre a proporção entre a altura e o diâmetro para a lata mais econômica (ou seja, que requer a menor quantidade de material para fabricação).


$$
V = \pi r^2 h
$$
$$
h = \frac{V}{\pi r^2}
$$
$$
A = 2 \pi r \left(\frac{V}{\pi r^2}\right) + 8r^2
$$
$$
A = \frac{2V}{r} + 8r^2
$$
$$
\frac{dA}{dr} = -\frac{2V}{r^2} + 16r = 0
$$
$$
16r^3 = 2V \implies r^3 = \frac{V}{8} \implies r = \left(\frac{V}{8}\right)^{\frac{1}{3}}
$$
$$
h = \frac{V}{\pi \left(\frac{V}{8}\right)^{\frac{2}{3}}} = \frac{8^{2/3}V}{\pi V^{2/3}} = \frac{8^{2/3}}{\pi} V^{1/3}
$$
$$
\frac{h}{2r} = \frac{\frac{8^{2/3}}{\pi} V^{1/3}}{2 \left(\frac{V}{8}\right)^{\frac{1}{3}}} = \frac{8^{2/3}}{2 \cdot 8^{1/3} \pi} = \frac{1}{\sqrt[3]{2} \pi}
$$

Portato:
$$
\frac{h}{2r} = \frac{1}{\sqrt[3]{2} \pi}
$$
