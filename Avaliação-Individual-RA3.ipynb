# Questão 1 - Vale até 1 ponto
Em uma região desértica, pesquisadores estão estudando o crescimento de uma espécie rara de cacto. O crescimento do cacto ao longo do tempo é modelado pela função $y = x^5(2 - x)$, onde $y$ representa a altura do cacto em metros e $x$ representa o tempo em anos desde o início do estudo. Os pesquisadores querem determinar a quantidade total de biomassa acumulada pelo cacto ao longo dos primeiros 2 anos. Sabemos, de estudos anteriores que a biomassa acumulada é diretamente proporcional (1:1) a área sob a curva de crescimento do cacto ao longo do tempo. Determine a quantidade total de biomassa acumulada nestes dois anos.



A integral a ser calculada é:

$$
\int_{0}^{2} x^5(2 - x) \, dx
$$

Expandindo a função, temos:

$$
\int_{0}^{2} (2x^5 - x^6) \, dx = \int_{0}^{2} 2x^5 \, dx - \int_{0}^{2} x^6 \, dx
$$

1) $\int_{0}^{2} 2x^5 \, dx$

$$
2 \int x^5 \, dx = 2 \left( \frac{x^6}{6} \right) = \frac{2x^6}{6} = \frac{x^6}{3}
$$

2) $\int_{0}^{2} x^6 \, dx$

$$
\int x^6 \, dx = \frac{x^7}{7}
$$

Substituindo na integral original:

$$
\int_{0}^{2} (2x^5 - x^6) \, dx = \left[ \frac{x^6}{3} \right]_{0}^{2} - \left[ \frac{x^7}{7} \right]_{0}^{2}
$$

Calculando os valores:

$\left[ \frac{x^6}{3} \right]_{0}^{2}$

$$
\left. \frac{x^6}{3} \right|_{0}^{2} = \frac{2^6}{3} - \frac{0^6}{3} = \frac{64}{3} - 0 = \frac{64}{3}
$$

$\left[ \frac{x^7}{7} \right]_{0}^{2}$

$$
\left. \frac{x^7}{7} \right|_{0}^{2} = \frac{2^7}{7} - \frac{0^7}{7} = \frac{128}{7} - 0 = \frac{128}{7}
$$

Agora subtraímos esses resultados:

$$
\int_{0}^{2} (2x^5 - x^6) \, dx = \frac{64}{3} - \frac{128}{7}
$$

Denominador comum:

$$
\frac{64}{3} = \frac{64 \cdot 7}{3 \cdot 7} = \frac{448}{21}
$$

$$
\frac{128}{7} = \frac{128 \cdot 3}{7 \cdot 3} = \frac{384}{21}
$$

Subtraindo:

$$
\frac{448}{21} - \frac{384}{21} = \frac{448 - 384}{21} = \frac{64}{21}
$$

Portanto, a área sob a curva $ y = x^5(2 - x) $ ao longo dos primeiros 2 anos é:

$$
\frac{64}{21}
$$

# Questão 2 - Vale até 1 ponto
Usando o Python (matplotlib, sympy e numpy) trace o gráfico e calcule a área sob a função $F(x) = x^5(2 - x)$ para $0 \leq x \leq 2$. Não esqueça de anotar os pontos, hachurar a área calculada e de imprimir o valor da área encontrada.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definindo a função
x = sp.symbols('x')
F = x**5 * (2 - x)

# Calculando a área sob a curva usando integração
area = sp.integrate(F, (x, 0, 2))
area_value = area.evalf()

# Gerando os pontos para o gráfico
x_vals = np.linspace(0, 2, 400)
F_func = sp.lambdify(x, F, 'numpy')
y_vals = F_func(x_vals)

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='$F(x) = x^5(2 - x)$', color='blue')
plt.fill_between(x_vals, y_vals, where=[(0 <= val <= 2) for val in x_vals], color='skyblue', alpha=0.4, hatch='//')
plt.title('Área sob a curva $F(x) = x^5(2 - x)$')
plt.xlabel('$x$')
plt.ylabel('$F(x)$')
plt.legend()
plt.grid(True)

# Anotando os pontos e a área
plt.annotate(f'Área = {area_value:.4f}', xy=(1, 5), fontsize=12, color='black')
plt.show()

# Imprimindo o valor da área encontrada
area_value
```

A aceleração de um objeto é dada por $a(t) = \cos(\pi t)$, e sua velocidade no tempo $t = 0$ é $\frac{1}{2\pi}$. Usando apenas álgebra, encontre a distância total percorrida por este objeto nos primeiros $1,5$ segundos deste movimento. Dica: antes de calcular qualquer coisa verifique cuidadosamente a equação da aceleração. Não é necessário, mas talvez seja interessante, traçar um gráfico da aceleração e outro da velocidade para entender o problema. É necessário entender o movimento do objeto para saber quanto espaço ele realmente percorreu.

Para determinar a distância total percorrida pelo objeto nos primeiros $1.5$ segundos, primeiro encontramos a função da velocidade:

$$
v(t) = \int a(t) \, dt
$$
$$
v(t) = \int \cos(\pi t) \, dt
$$

Integrando, temos:

$$
v(t) = \frac{\sin(\pi t)}{\pi} + C  
$$

Sabendo que quando $ t = 0 $, $ v(0) = \frac {1}{2\pi} $, podemos comparar com $ v(t) $ para encontrar o valor de $ C $:

$$
\frac{\sin(0)}{\pi} + C = \frac {1}{2\pi}
$$

$$
C = \frac {1}{2\pi}
$$

Portanto, a função da velocidade é:

$$
v(t) = \frac{\sin(\pi t )}{\pi}  + \frac {1}{2\pi}
$$

Agora, encontramos a função de posição:

$$
s(t) = \int \left( \frac{\sin(\pi t )}{\pi}  + \frac {1}{2\pi} \right) dt
$$

Integrando e repetindo o mesmo processo para descobrir $ C $, temos:

$$
s(t) = - \frac{\cos(\pi t )}{\pi^2}  + \frac {t}{2\pi} + C
$$

Sabendo que $ s(0) = 0 $:

$$
- \frac{\cos(0 )}{\pi^2}  + \frac {0}{2\pi} + C = 0
$$

$$
- \frac{1}{\pi^2} = -C
$$

$$
C = \frac{1}{\pi^2}
$$

Portanto, a função de posição é:

$$
s(t) = - \frac{\cos(\pi t )}{\pi^2}  + \frac {t}{2\pi} + \frac{1}{\pi^2}
$$

Finalmente, calculamos a distância total percorrida em 1.5 segundos:

$$
\text{Distância} = s(1.5) - s(0)
$$

Para $ t = 1.5 $:

$$
s(1.5) = - \frac{\cos(\pi \cdot 1.5 )}{\pi^2}  + \frac {1.5}{2\pi} + \frac{1}{\pi^2}
$$

Como $\cos(\pi \cdot 1.5) = 0$:

$$
s(1.5) = \frac {0}{\pi^2} + \frac {1.5}{2\pi} + \frac{1}{\pi^2}
$$

Simplificando:

$$
s(1.5) = \frac{1.5}{2\pi} + \frac{1}{\pi^2} = 0.34
$$

Para $ t = 0 $:

$$
s(0) = - \frac{\cos(\pi \cdot 0 )}{\pi^2}  + \frac {0}{2\pi} + \frac{1}{\pi^2}
$$

Como $\cos(\pi \cdot 0) = 1$:

$$
s(0) = - \frac{1}{\pi^2} + \frac{1}{\pi^2} = 0
$$

Portanto, a distância percorrida é $0.34$.

# Questão 4 - Vale até 1 ponto

Encontre a área limitada por $f(x) = -x^2 + 4x + 1$, $g(x) = -x^3 + 7x^2 - 10x + 3$ para  $1\leq x \leq 2$. Nesta questão você pode usar álgebra, ou Python, considerando as limitações de usos de bibliotecas definidas para esta avaliação.  


```python
import sympy as sp

x = sp.symbols('x')
f = -x**2 + 4*x + 1
g = -x**3 + 7*x**2 - 10*x + 3

difference_function = sp.Abs(f - g)
total_area = sp.integrate(difference_function, (x, 1, 2))

print("A área limitada pelas duas curvas é:", total_area.evalf())
```

Para determinar o volume do sólido obtido pela rotação da região delimitada pelas funções $ y = \sqrt{x} $, $ y = 3 $ e o eixo $ y $ em torno do eixo $ y $, usaremos o método dos discos.

Primeiro, precisamos expressar $ x $ como uma função de $ y $. A função $ y = \sqrt{x} $ pode ser invertida para obter:

$$ x = y^2 $$

A região de interesse é delimitada entre $ y = 0 $ e $ y = 3 $.

O volume $ V $ do sólido de revolução é dado pela fórmula:

$$
V = \pi \int_{a}^{b} [R(y)]^2 \, dy
$$

onde $ R(y) $ é o raio do disco em um dado $ y $. Neste caso, o raio $ R(y) $ é a distância entre a curva $ x = y^2 $ e o eixo $ y $, então $ R(y) = y^2 $.

Portanto, a integral para calcular o volume é:

$$
V = \pi \int_{0}^{3} (y^2)^2 \, dy = \pi \int_{0}^{3} y^4 \, dy
$$

Agora, vamos calcular esta integral:

$$
V = \pi \int_{0}^{3} y^4 \, dy
$$

Integrando:

$$
V = \pi \left[ \frac{y^5}{5} \right]_{0}^{3}
$$

Substituindo os limites de integração:

$$
V = \pi \left( \frac{3^5}{5} - \frac{0^5}{5} \right)
$$

$$
V = \pi \left( \frac{243}{5} \right)
$$

Portanto, o volume é:

$$
V = \frac{243\pi}{5}
$$

O volume do sólido obtido pela rotação da região delimitada pelas funções $ y = \sqrt{x} $, $ y = 3 $, e o eixo $ y $ em torno do eixo $ y $ é:

$$
\frac{243\pi}{5} = 152.68

Para os problemas a e b, determine a média, usando a integração, para a funções nos intervalos especificados.

a. $f(x) = 8x - 3 + 5e^{2-x}$ no intervalo $[0, 2]$ - Vale 1 ponto

b. Considere que a posição de um objeto em movimento ao longo do eixo $ x $ é descrita pela função de deslocamento $ x(t) = t^3 - 6t^2 + 9t + 1 $, onde $ x(t) $ está em metros e $t$ está em segundos. Determine a velocidade média do objeto no intervalo de tempo de $ t = 1$ segundo até $t = 4$ segundos. - Vale 2 pontos

### Problema A

Para determinar a média da função $ f(x) = 8x - 3 + 5e^{2-x} $ no intervalo $[0, 2]$, usamos a fórmula da média de uma função contínua $ f(x) $ no intervalo $[a, b]$:

$$
\text{Média} = \frac{1}{b-a} \int_{a}^{b} f(x) \, dx
$$

Aqui, $ a = 0 $ e $ b = 2 $.

Então, precisamos calcular:

$$
\text{Média} = \frac{1}{2-0} \int_{0}^{2} (8x - 3 + 5e^{2-x}) \, dx
$$

Primeiro, calculamos a integral:

$$
\int_{0}^{2} (8x - 3 + 5e^{2-x}) \, dx
$$

Podemos dividir a integral em partes:

$$
\int_{0}^{2} 8x \, dx - \int_{0}^{2} 3 \, dx + \int_{0}^{2} 5e^{2-x} \, dx
$$

Calculando cada uma das integrais separadamente:

1. Para $ \int_{0}^{2} 8x \, dx $:

$$
\int_{0}^{2} 8x \, dx = 8 \left[ \frac{x^2}{2} \right]_{0}^{2} = 8 \left( \frac{2^2}{2} - \frac{0^2}{2} \right) = 8 \times 2 = 16
$$

2. Para $ \int_{0}^{2} 3 \, dx $:

$$
\int_{0}^{2} 3 \, dx = 3 \left[ x \right]_{0}^{2} = 3 (2 - 0) = 6
$$

3. Para $ \int_{0}^{2} 5e^{2-x} \, dx $:

Usamos a substituição $ u = 2 - x $, então $ du = -dx $.

$$
\int_{0}^{2} 5e^{2-x} \, dx = 5 \int_{2}^{0} e^u (-du) = 5 \int_{0}^{2} e^u \, du = 5 [e^u]_{0}^{2} = 5 (e^2 - e^0) = 5 (e^2 - 1)
$$

Somando os resultados:

$$
\int_{0}^{2} (8x - 3 + 5e^{2-x}) \, dx = 16 - 6 + 5(e^2 - 1) = 10 + 5e^2 - 5 = 5e^2 + 5
$$

Agora, calculamos a média:

$$
\text{Média} = \frac{1}{2} (5e^2 + 5) = \frac{5e^2 + 5}{2}
$$

Portanto, a média da função $ f(x) = 8x - 3 + 5e^{2-x} $ no intervalo $[0, 2]$ é:

$$
\frac{5e^2 + 5}{2}
$$

# Questão 7 - Vale até 2 pontos

Nos ítens a, b a seguir, encontre a família de antiderivadas das seguintes funções:

a. $f(x) = 4x^3 - 6x^2 + 2x - 5 $.

b. $f(x) = \frac{1}{x} + e^x$ para todo $x\neq 0$.




### Problema A

Para encontrar a família de antiderivadas da função $ f(x) = 4x^3 - 6x^2 + 2x - 5 $, integramos cada termo separadamente.

1. Para $ 4x^3 $:

$$
\int 4x^3 \, dx = 4 \int x^3 \, dx = 4 \left( \frac{x^4}{4} \right) = x^4
$$

2. Para $ -6x^2 $:

$$
\int -6x^2 \, dx = -6 \int x^2 \, dx = -6 \left( \frac{x^3}{3} \right) = -2x^3
$$

3. Para $ 2x $:

$$
\int 2x \, dx = 2 \int x \, dx = 2 \left( \frac{x^2}{2} \right) = x^2
$$

4. Para $ -5 $:

$$
\int -5 \, dx = -5x
$$

Somando todas as antiderivadas e adicionando a constante de integração $ C $:

$$
\int (4x^3 - 6x^2 + 2x - 5) \, dx = x^4 - 2x^3 + x^2 - 5x + C
$$

Portanto, a família de antiderivadas de $ f(x) = 4x^3 - 6x^2 + 2x - 5 $ é:

$$
x^4 - 2x^3 + x^2 - 5x + C
$$

### Problema B

Para encontrar a família de antiderivadas da função $ f(x) = \frac{1}{x} + e^x $, integramos cada termo separadamente.

1. Para $ \frac{1}{x} $:

$$
\int \frac{1}{x} \, dx = \ln|x|
$$

2. Para $ e^x $:

$$
\int e^x \, dx = e^x
$$

Somando ambas as antiderivadas e adicionando a constante de integração $ C $:

$$
\int \left( \frac{1}{x} + e^x \right) \, dx = \ln|x| + e^x + C
$$

Portanto, a família de antiderivadas de $ f(x) = \frac{1}{x} + e^x $ é:

$$
\ln|x| + e^x + C
$$
